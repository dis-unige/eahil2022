```python
import re
import os
import pandas as pd
import time
import datetime
```

```python
# prefix for the files exported
file_prefix = 'UNIGE_WoS_2001_2020'

# change working directory
working_directory = 'data/sources/'

# read file names with txt extension
list_of_txt_files = []
for file in os.listdir(working_directory):
    if file.endswith(".txt"):
        list_of_txt_files.append(file)
        
# Sort Strings on numerical substrings
list_of_txt_files.sort(key=lambda list_of_txt_files : list(map(int, re.findall(r'\d+', list_of_txt_files)))[0])

# loop files in the directory
list_of_dfs = []
for file in list_of_txt_files:
    
    # read file
    target_df = pd.read_csv('data/sources/' + file, sep='\t', lineterminator='\r', index_col=False, skip_blank_lines=True)
    
    # remove nan rows
    # target_df = target_df.dropna(how = "all")
    target_df.drop(target_df.tail(1).index,inplace=True)
    
    # print summary
    print("current file: ", "\t","\t", file)
    print("number of columns: ", "\t", target_df.shape[1])
    print("number of rows: ", "\t", target_df.shape[0])
    print("")
    
    # add df to list of df
    list_of_dfs.append(target_df)
    
del(file); del(target_df); del(list_of_txt_files)    
    
# concatenate all df
df_concatenated = pd.concat(list_of_dfs)
del(list_of_dfs)

# save file
# df_concatenated.to_csv('big_WoS_df.csv', index=False, header=True, sep = ";")
# df_concatenated.to_excel('big_WoS_df.xlsx', index=False, header=True)
```

```python
def create_dataframe(df1):
    
    # Get all the data from the columns
    column_CR = list(df1["CR"])
    column_WOS = list(df1["UT"])
    column_PY = list(df1["PY"])
    length_df_rows = df1.shape[0]
    
    # Create variable for all dataframe rows
    dataframe_rows = []
    
    # Add columns
    dataframe_rows.append(['WoS accession of article', 
                           'Publication Year of article', 
                           'Author of cited article', 
                           'Year of cited article', 
                           'Journal of cited article',
                           'DOI',
                           'Years difference', 
                           'Need to be checked'])
    
    # loop all df rows
    for x in range(0, length_df_rows):
        
        # Wos unique code
        wos_code = column_WOS[x]
        try:
            wos_code.lstrip()
        except:
            pass
            
        # Publication year
        publication_year = column_PY[x]
        
        try:
            publication_year = int(publication_year)
        except:
            pass
            
        # cited articles
        cited_articles = column_CR[x]
        
        try:
            cited_articles = column_CR[x]
            cited_articles = cited_articles.split(";")
            
            # loop all cited articles
            for article in cited_articles:
                try:
                    article_list = article.split(",")
                    cited_article_author = article_list[0].lstrip()
                    cited_article_year = int(article_list[1].lstrip())
                    cited_article_journal = article_list[2].lstrip()
                    cited_article_doi = article_list[-1].lstrip()
                    row = [wos_code, publication_year, cited_article_author, cited_article_year, cited_article_journal, cited_article_doi, publication_year - cited_article_year, "No"]
                    dataframe_rows.append(row)
                    
                except:
                    dataframe_rows.append([wos_code, publication_year, article, "", "", "", "", "Yes"])
        except:
            dataframe_rows.append([wos_code, publication_year, "", "", "", "", "", "Yes"])
            
    df_final = pd.DataFrame(dataframe_rows[1:], columns = dataframe_rows[0])
    return(df_final)
```
